## 问题 1  

> 给定一个已排序的链表的头 `head` ， *删除所有重复的元素，使每个元素只出现一次* 。返回 *已排序的链表* 。链表的类如下所示：

```python
class NodeList:
    def __init__(self, val=None, right=None):
        self.val   = val
        self.right = right
```

输入是一个数组，你首先需要将数组转化为链表，然后删除链表中的重复元素，再遍历链表元素，以一个数组的形式返回。请设计一个算法解决上述任务，分析算法设计思路，计算时间复杂度, 并基于python编程实现。

e.g.  输入：head=[1, 1, 2, 3, 3]   输出：[1, 2, 3]

![image-20240502110020439](./fig/hw2q1.png)



idea：
### 算法设计思路
1. **数组转链表**：遍历输入数组，逐个创建链表节点，并将它们连接起来，形成一个链表。
2. **删除重复元素**：使用一个指针遍历链表，比较当前节点和下一个节点的值。如果值相同，则跳过下一个节点，直接将当前节点的`right`指针指向下下个节点；如果值不同，则将指针移动到下一个节点。
3. **链表转数组**：遍历处理后的链表，将每个节点的值添加到一个数组中，最后返回该数组。

### 时间复杂度分析
- **数组转链表**：需要遍历一次输入数组，时间复杂度为 $O(n)$，其中 $n$ 是数组的长度。
- **删除重复元素**：只需要遍历一次链表，时间复杂度为 $O(n)$。
- **链表转数组**：同样需要遍历一次链表，时间复杂度为 $O(n)$。

总体时间复杂度为 $O(n)$，因为每个操作都只需要遍历一次链表或数组。


````python
# add your idea here
class NodeList:
    def __init__(self, val=None, right=None):
        # 节点的值
        self.val = val
        # 指向下一个节点的指针
        self.right = right

def array_to_linked_list(arr):
    # 创建一个虚拟头节点，方便操作
    dummy = NodeList()
    # 当前节点指针，初始指向虚拟头节点
    current = dummy
    # 遍历输入数组
    for val in arr:
        # 创建一个新节点，值为当前数组元素
        current.right = NodeList(val)
        # 将当前节点指针移动到新节点
        current = current.right
    # 返回虚拟头节点的下一个节点，即链表的头节点
    return dummy.right

def remove_duplicates(head):
    # 当前节点指针，初始指向链表头节点
    current = head
    # 当当前节点和下一个节点都存在时
    while current and current.right:
        # 如果当前节点的值和下一个节点的值相同
        if current.val == current.right.val:
            # 跳过下一个节点，直接将当前节点的 right 指针指向下下个节点
            current.right = current.right.right
        else:
            # 如果值不同，将当前节点指针移动到下一个节点
            current = current.right
    # 返回处理后的链表头节点
    return head

def linked_list_to_array(head):
    # 用于存储链表节点值的数组
    result = []
    # 当前节点指针，初始指向链表头节点
    current = head
    # 遍历链表
    while current:
        # 将当前节点的值添加到结果数组中
        result.append(current.val)
        # 将当前节点指针移动到下一个节点
        current = current.right
    # 返回结果数组
    return result

def solution(head):
    # 将数组转换为链表
    linked_list = array_to_linked_list(head)
    # 删除链表中的重复元素
    new_linked_list = remove_duplicates(linked_list)
    # 将链表转换为数组
    return linked_list_to_array(new_linked_list)

# 测试代码
head = [1, 1, 2, 3, 3]
print(solution(head))
    
# your algorithm time complexity is: O(n)
 
````

## 问题 2  

> 下面是一个经典的算法问题：
>
> - 给定包含n个整数的一个整数数组 `nums` 和一个整数目标值 `target`，请你在该数组中找出 **和为目标值** *`target`* 的那 **两个** 整数，并返回它们的**数组下标**。假设每种输入只会对应一个答案。但是，数组中同一个元素在答案里不能重复出现。你可以按任意顺序返回答案。
>
> 由于要多次查找数组中元素的位置，为了提高查询效率可以使用哈希表来存储数组中的数据，在哈希表中查询一个元素的复杂度为O(1)。 已知python中的字典是使用哈希表实现的，即使用`dict[key]`查询对应的value时间复杂度为O(1), python提供了查询字典是否包含某个key的功能：`key in dict`，时间复杂度也是O(1)

请根据上面信息，设计一个时间复杂度为O(n) 的算法，解决上述算法问题

e.g.   

输入：nums=[2,7,11,15], target=9， 输出：[0，1]

输入：nums=[3,2,4], target=6, 输出：[1,2]

输入：nums=[3,3], target=6,  输出：[0,1]


idea:

### 算法思路
1. 初始化一个空字典 `num_dict`，用于存储数组元素及其对应的下标。
2. 遍历数组 `nums`，对于每个元素 `num` 及其下标 `i`：
    - 计算 `target` 与当前元素 `num` 的差值 `complement`。
    - 检查 `complement` 是否在 `num_dict` 中。
        - 如果存在，说明找到了两个数之和等于 `target`，返回 `num_dict[complement]` 和当前下标 `i`。
        - 如果不存在，将当前元素 `num` 及其下标 `i` 存入 `num_dict` 中。

### 复杂度分析
- **时间复杂度**：只需要遍历一次数组，每次在字典中查找和插入操作的时间复杂度都是 $O(1)$，所以总的时间复杂度为 $O(n)$，其中 $n$ 是数组的长度。
- **空间复杂度**：最坏情况下，需要存储数组中的所有元素，所以空间复杂度为 $O(n)$。



````python
# add your idea here
def twoSum(nums, target):
    # 初始化一个空字典，用于存储数组元素及其下标
    num_dict = {}
    # 遍历数组，同时获取元素及其下标
    for i, num in enumerate(nums):
        # 计算目标值与当前元素的差值
        complement = target - num
        # 检查差值是否在字典中
        if complement in num_dict:
            # 如果存在，返回差值的下标和当前元素的下标
            return [num_dict[complement], i]
        # 将当前元素及其下标存入字典
        num_dict[num] = i

# 测试示例
nums1 = [2, 7, 11, 15]
target1 = 9
print(twoSum(nums1, target1))

nums2 = [3, 2, 4]
target2 = 6
print(twoSum(nums2, target2))

nums3 = [3, 3]
target3 = 6
print(twoSum(nums3, target3))
    
# your algorithm time complexity is: O(n)
````

## 问题 3:   

> 栈是一种常用的数据结构，编译器中通常使用栈来实现表达式求值。
>
> 以表达式 $3+5 \times 8-6$​ 为例。编译器使用两个栈来完成运算，即一个栈保持操作数，另一个栈保存运算符。
>
> 1. 从左向右遍历表达式，遇到数字就压入操作数栈；
>
> 2. 遇到运算符，就与运算符栈的栈顶元素进行比较。如果比运算符栈顶元素的优先级高，就将当前运算符压入栈；如果比运算符栈顶元素的优先级低或者相同，从运算符栈中取栈顶运算符，从操作数栈的栈顶取 2 个操作数，然后进行计算，再把计算完的结果压入操作数栈，继续比较。
>
> 下图是 $3+5 \times 8-6$  这个表达式的计算过程：

![figure](./fig/hw2q3.png)

根据上述原理，请设计一个算法完成表达式的运算，当输入为表达式字符串，返回对应的计算结果。分析算法设计思路，计算时间复杂度，并基于python编程实现

**note：**

1. 假设输入的表达式只会出现加（“+”），减 “-”， 乘“*”，除 “/” 四个运算符, 表达式中只会出现正整数
2. python中` str.isdigit()`函数可以判断字符串str是否为数字，



e.g. :
---

1. 输入：“$3+5 * 8 -6$”   输出：37

2. 输入：“$34+13*9 + 44-12/3$”  输出：191

idea：
### 算法设计思路
1. **初始化栈**：创建两个栈，一个用于存储操作数（`operand_stack`），另一个用于存储运算符（`operator_stack`）。
2. **定义运算符优先级**：为加、减、乘、除运算符定义优先级，其中乘除的优先级高于加减。
3. **遍历表达式**：从左到右遍历输入的表达式字符串。
    - **处理数字**：如果遇到数字，将其连续读取并转换为整数，然后压入操作数栈。
    - **处理运算符**：如果遇到运算符，将其与运算符栈的栈顶元素进行比较。
        - 若当前运算符的优先级高于栈顶运算符，将当前运算符压入运算符栈。
        - 若当前运算符的优先级低于或等于栈顶运算符，从运算符栈中弹出栈顶运算符，从操作数栈中弹出两个操作数进行计算，将计算结果压入操作数栈，然后继续比较当前运算符与新的栈顶运算符。
4. **处理剩余运算符**：遍历结束后，若运算符栈中还有剩余的运算符，依次弹出运算符和操作数进行计算，直到运算符栈为空。
5. **返回结果**：操作数栈中最终剩下的元素即为表达式的计算结果。

### 时间复杂度分析
- 遍历表达式字符串一次，时间复杂度为 $O(n)$，其中 $n$ 是表达式字符串的长度。
- 在遍历过程中，每个运算符和操作数最多入栈和出栈一次，因此栈操作的总时间复杂度也是 $O(n)$。

总体时间复杂度为 $O(n)$。

````python
# add your idea here
def apply_operator(operators, operands):
    # 从运算符栈中弹出栈顶运算符
    operator = operators.pop()
    # 从操作数栈中弹出两个操作数
    right_operand = operands.pop()
    left_operand = operands.pop()
    if operator == '+':
        operands.append(left_operand + right_operand)
    elif operator == '-':
        operands.append(left_operand - right_operand)
    elif operator == '*':
        operands.append(left_operand * right_operand)
    elif operator == '/':
        operands.append(left_operand / right_operand)

def precedence(operator):
    # 定义运算符优先级
    if operator in ('+', '-'):
        return 1
    if operator in ('*', '/'):
        return 2
    return 0

def evaluate_expression(expression):
    # 初始化操作数栈和运算符栈
    operands = []
    operators = []
    i = 0
    while i < len(expression):
        if expression[i].isdigit():
            num = 0
            # 连续读取数字
            while i < len(expression) and expression[i].isdigit():
                num = num * 10 + int(expression[i])
                i += 1
            # 将数字压入操作数栈
            operands.append(num)
            i -= 1
        elif expression[i] in ('+', '-', '*', '/'):
            # 处理运算符
            while (operators and precedence(operators[-1]) >= precedence(expression[i])):
                apply_operator(operators, operands)
            # 将当前运算符压入运算符栈
            operators.append(expression[i])
        i += 1
    # 处理剩余的运算符
    while operators:
        apply_operator(operators, operands)
    # 返回最终结果
    return operands[-1]

# 测试示例
expression1 = "3+5*8-6"
print(evaluate_expression(expression1))

expression2 = "34+13*9+44-12/3"
print(evaluate_expression(expression2))
    
# your algorithm time complexity is: O（n）
````

## 问题 4:  

> 星球碰撞问题：现有n个星球，在同一条直线上运行，如数组A所示，元素的绝对值表示星球的质量，负数表示星球自右向左运动，正数表示星球自左向右运动，当两个星球相撞的时候，质量小的会消失，大的保持不变，**质量相同的两个星球碰撞后自右向左运动的星球消失，自左向右的星球保持不变**，假设所有星球的速度大小相同。
>
> $ A=[23,-8, 9, -3, -7, 9, -23, 22] $

请设计一个算法模拟星球的运行情况，输出最终的星球存续情况（输出一个数组），分析算法设计思路，计算时间复杂度，并基于python编程实现。

e.g.
---
1.  输入： A=[-3,-6,2,8, 5,-8,9,-2,1]， 输出：[-3, -6, 2, 8, 9, 1]

2. 输入：A=[23,-8, 9, -3, -7, 9, -23, 22], 输出：[23, 22]



idea:

### 算法设计思路
1. 初始化一个空栈 `stack` 用于存储当前存活的星球。
2. 遍历数组 `A` 中的每个星球：
    - 如果栈为空，直接将当前星球压入栈中。
    - 如果栈不为空：
        - 取出栈顶的星球，记为 `top`。
        - 如果 `top` 和当前星球的运动方向相同（即 `top` 和当前星球都为正或者都为负），则将当前星球压入栈中。
        - 如果 `top` 和当前星球的运动方向相反（一个为正一个为负）：
            - 比较 `top` 和当前星球的质量绝对值大小。
            - 如果 `top` 的质量绝对值大于当前星球的质量绝对值，不做任何操作，因为当前星球会消失。
            - 如果 `top` 的质量绝对值小于当前星球的质量绝对值，弹出栈顶元素 `top`，然后递归地将当前星球与新的栈顶元素（如果栈不为空）进行碰撞判断，直到栈为空或者栈顶元素与当前星球运动方向相同。
            - 如果 `top` 的质量绝对值等于当前星球的质量绝对值，且 `top` 为负（自右向左运动），弹出栈顶元素 `top`，因为自右向左运动的星球会消失；若 `top` 为正（自左向右运动），则将当前星球丢弃，因为自左向右运动的星球会保持不变。
3. 遍历结束后，栈中剩下的元素就是最终存活的星球，返回栈中元素组成的数组。

### 时间复杂度分析
对于数组 `A` 中的每个元素，最多进行一次入栈和出栈操作，而栈操作的时间复杂度为 $O(1)$。所以，遍历数组的时间复杂度为 $O(n)$，其中 $n$ 是数组 `A` 的长度。因此，总的时间复杂度为 $O(n)$。

````python
def simulate_planets(A):
    stack = []
    for planet in A:
        while stack and (stack[-1] > 0) != (planet > 0):
            if abs(stack[-1]) > abs(planet):
                break
            elif abs(stack[-1]) < abs(planet):
                stack.pop()
            else:
                if stack[-1] < 0:
                    stack.pop()
                break
        else:
            stack.append(planet)
    return stack

# 测试示例
A1 = [-3, -6, 2, 8, 5, -8, 9, -2, 1]
print(simulate_planets(A1))

A2 = [23, -8, 9, -3, -7, 9, -23, 22]
print(simulate_planets(A2))
````

## 问题 5 

> 给定一个无序数组nums=[9,-3,-10,0,9,7,33]，请建立一个二叉搜索树存储数组中的所有元素，之后删除二叉树中的元素“0”，再使用中序遍历输出二叉搜索树中的所有元素。

使用python编程完成上述任务，并计算时间复杂度


idea:

### 算法思路
1. **构建二叉搜索树**：二叉搜索树（BST）的特点是左子树的所有节点值小于根节点值，右子树的所有节点值大于根节点值。遍历数组 `nums`，将每个元素依次插入到 BST 中。
2. **删除节点**：在 BST 中找到值为 `0` 的节点并删除。删除操作分三种情况：
    - 若节点是叶子节点，直接删除。
    - 若节点只有一个子节点，用该子节点替换要删除的节点。
    - 若节点有两个子节点，找到右子树中的最小节点（即右子树中最左边的节点），用该最小节点的值替换要删除节点的值，然后删除右子树中的最小节点。
3. **中序遍历**：中序遍历 BST 会按升序输出节点值。递归遍历左子树，访问根节点，再递归遍历右子树。

### 复杂度分析
- **构建二叉搜索树**：插入一个节点的时间复杂度为 $O(h)$，其中 $h$ 是树的高度。在最坏情况下，树退化为链表，高度为 $n$（$n$ 是数组长度），插入 $n$ 个节点的总时间复杂度为 $O(n^2)$；在平均情况下，树是平衡的，高度为 $O(\log n)$，总时间复杂度为 $O(n \log n)$。
- **删除节点**：删除操作的时间复杂度同样为 $O(h)$，最坏情况为 $O(n)$，平均情况为 $O(\log n)$。
- **中序遍历**：遍历每个节点一次，时间复杂度为 $O(n)$。

整个过程的时间复杂度在最坏情况下为 $O(n^2)$，平均情况下为 $O(n \log n)$。 

````python
# add your idea here
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

# 插入节点到二叉搜索树
def insert(root, val):
    if root is None:
        return TreeNode(val)
    if val < root.val:
        root.left = insert(root.left, val)
    else:
        root.right = insert(root.right, val)
    return root

# 找到二叉搜索树中的最小节点
def find_min(node):
    while node.left:
        node = node.left
    return node

# 删除二叉搜索树中的节点
def delete_node(root, key):
    if root is None:
        return root
    if key < root.val:
        root.left = delete_node(root.left, key)
    elif key > root.val:
        root.right = delete_node(root.right, key)
    else:
        # 情况 1: 没有子节点或只有一个子节点
        if root.left is None:
            return root.right
        elif root.right is None:
            return root.left
        # 情况 2: 有两个子节点
        temp = find_min(root.right)
        root.val = temp.val
        root.right = delete_node(root.right, temp.val)
    return root

# 中序遍历二叉搜索树
def inorder_traversal(root):
    result = []
    if root:
        result = inorder_traversal(root.left)
        result.append(root.val)
        result = result + inorder_traversal(root.right)
    return result

# 主函数
def main():
    nums = [9, -3, -10, 0, 9, 7, 33]
    root = None
    # 构建二叉搜索树
    for num in nums:
        root = insert(root, num)
    # 删除值为 0 的节点
    root = delete_node(root, 0)
    # 中序遍历并输出结果
    result = inorder_traversal(root)
    print(result)

if __name__ == "__main__":
    main()
    
# your algorithm time complexity is:平均情况：O(nlogn),最坏情况：O(n*n)
````

## 问题 6  

> 给定一个包含大写字母和小写字母的字符串 s ，返回 字符串包含的 **最长的回文子串的长度** 。请注意 区分大小写 。比如 "Aa" 不能当做一个回文字符串。

请设计一个算法解决上述问题，只需要输出最长回文子串的长度，分析算法设计思路，计算时间复杂度，并基于python编程实现

e.g. 输入： s="adccaccd"，  输出：7。 最长回文子串为："dccaccd", 长度为7


idea：
### 算法设计思路
1. **暴力枚举法**：枚举字符串 `s` 的所有子串，判断每个子串是否为回文串，并记录最长回文子串的长度。
    - 枚举所有可能的子串起始位置 `i` 和结束位置 `j`（其中 `0 <= i <= j < len(s)`）。
    - 对于每个子串 `s[i:j+1]`，判断其是否为回文串，即判断该子串正序和逆序是否相等。
    - 记录满足回文条件的子串中长度最大的那个子串的长度。
2. **中心扩展法**：以字符串中的每个字符为中心，向两边扩展，判断扩展出的子串是否为回文串。由于回文串的中心可以是一个字符或者两个字符（例如 “aba” 中心是 “a”，“abba” 中心是 “bb”），所以需要分别考虑这两种情况。
    - 遍历字符串 `s` 的每个字符，分别以该字符为中心向两边扩展。
    - 对于每个字符，分别考虑以它为单字符中心和双字符中心（如果存在相邻相同字符）的情况，向两边扩展直到不再是回文串为止。
    - 记录扩展过程中得到的最长回文子串的长度。

### 时间复杂度分析
1. **暴力枚举法**：
    - 枚举子串的起始位置 `i` 有 `n` 种可能，枚举结束位置 `j` 有 `n` 种可能（`n` 为字符串 `s` 的长度），所以枚举子串的时间复杂度为 $O(n^2)$。
    - 对于每个子串，判断其是否为回文串的时间复杂度为 $O(n)$（因为需要比较子串的字符）。
    - 总的时间复杂度为 $O(n^3)$。
2. **中心扩展法**：
    - 字符串中有 `n` 个字符，对于每个字符，最多扩展 `n` 次（极端情况是整个字符串都是回文串）。
    - 考虑单字符中心和双字符中心两种情况，所以总的时间复杂度为 $O(n^2)$。

````python
# add your idea here
def longest_palindrome(s):
    n = len(s)
    max_len = 0
    for i in range(n):
        # 以s[i]为中心的回文串
        len1 = expand_around_center(s, i, i)
        # 以s[i]和s[i+1]为中心的回文串（如果存在s[i+1]且s[i] == s[i+1]）
        len2 = expand_around_center(s, i, i + 1) if i < n - 1 and s[i] == s[i + 1] else 0
        max_len = max(max_len, len1, len2)
    return max_len

def expand_around_center(s, left, right):
    while left >= 0 and right < len(s) and s[left] == s[right]:
        left -= 1
        right += 1
    return right - left - 1

# 测试示例
s = "adccaccd"
print(longest_palindrome(s))
# your algorithm time complexity is:O(n*n)
````

## 问题 7 

> 沿一条长河流分散着n座房子。你可以把这条河想象成一条轴，房子是由它们在这条轴上的坐标按顺序排列的。你的公司想在河边的特定地点设置手机基站，这样每户人家都在距离基站4公里的范围内。输入可以看作为一个升序数组，数组元素的取值为大于等于0的正整数，你需要输出最小基站的数目，基站的位置。

1. 给出一个时间复杂度为$O(n$) 的算法，使所使用的基站数量最小化，分析算法设计思路，使用python编程实现
2. 证明1.中算法产生了最优解决方案。

e.g. 

输入： [1, 5, 12, 33, 34,35]  输出：基站数目为3， 基站位置为[1，12，33]



idea：

### 算法思路：
1. 从第一个房子开始，把第一个房子的位置当作第一个基站的位置，同时记录下这个基站的覆盖范围的右边界（即房子位置加上 4 ）。
2. 遍历房子列表，检查下一个房子是否在当前基站的覆盖范围内，如果在，则继续检查下一个房子，直到找到一个不在覆盖范围内的房子。
3. 当找到一个不在覆盖范围内的房子时，将这个房子的位置设为新的基站位置，同时更新覆盖范围的右边界。
4. 重复步骤 2 和 3 ，直到遍历完所有的房子。
5. 最后返回基站的数量和基站的位置列表。

### 时间复杂度分析：
通过一个 while 循环遍历了一次输入的房子列表，对于列表中的每个元素最多进行一次检查和处理，因此时间复杂度为 \(O(n)\)，其中 \(n\) 是房子的数量。 




### 最优性证明
假设存在一个最优解 $S_{opt}$，其使用的基站数量为 $m_{opt}$，我们的贪心算法得到的解为 $S_{greedy}$，使用的基站数量为 $m_{greedy}$。
- **贪心选择性质**：在每一步，贪心算法总是选择在当前未覆盖的房子中，能覆盖尽可能多后续房子的位置设置基站。也就是说，对于第一个未被覆盖的房子 $h_i$，我们选择距离它不超过 4 公里范围内最右边的房子 $h_j$ 处设置基站，这样可以覆盖从 $h_i$ 到 $h_j$ 以及后续尽可能多的房子。
- **最优子结构性质**：当我们在 $h_j$ 处设置基站后，问题就转化为覆盖剩余未被覆盖的房子的子问题。由于贪心算法在每一步都做出了最优选择，所以子问题的最优解可以组合成原问题的最优解。
- **反证法**：假设 $m_{greedy} > m_{opt}$，即贪心算法使用的基站数量比最优解多。但是，由于贪心算法每次都尽可能地覆盖更多的房子，所以不可能存在一个比贪心算法使用更少基站的解，这与假设矛盾。因此，$m_{greedy} \leq m_{opt}$。又因为 $S_{opt}$ 是最优解，所以 $m_{greedy} = m_{opt}$，即贪心算法得到的解是最优解。

````python
# add your idea here
def min_base_stations(houses):
    base_station_count = 0
    base_station_positions = []
    i = 0
    while i < len(houses):
        base_station_count += 1
        base_station_positions.append(houses[i])
        coverage_end = houses[i] + 4
        while i + 1 < len(houses) and houses[i + 1] <= coverage_end:
            i += 1
        i += 1
    return base_station_count, base_station_positions


houses = [1, 5, 12, 33, 34, 35]
count, positions = min_base_stations(houses)
print(f"基站数目为{count}， 基站位置为{positions}")

    
# your algorithm time complexity is: 0(n)
````

## 问题 8  

> 给定由n个正整数组成的一个集合$S = \{a_1, a_2，···，a_n\}$和一个正整数W，设计一个算法确定是否存在S的一个子集 $K \subseteq S$, 使K中所有数之和为 $W$, 如果存在返回“True”，否则返回“False”

请设计一个时间复杂度为$O(nW)$动态规划算法，解决上述问题，分析算法的设计思路，并且基于python编程实现（不需要输出子集）。

e.g. 

输入：S = {1,4,7,3,5}， W = 11，输出：True。   因为K可以是{4,7}。



idea：
### 算法设计思路
使用动态规划来解决子集和问题。动态规划的核心思想是将大问题分解为小问题，并保存小问题的解，避免重复计算。

设 `dp[i][w]` 表示集合 `S` 的前 `i` 个元素是否存在一个子集，其元素之和为 `w`。那么有以下两种情况：
1. **不选择第 `i` 个元素**：若前 `i - 1` 个元素中已经存在一个子集的和为 `w`，即 `dp[i - 1][w]` 为 `True`，则 `dp[i][w]` 也为 `True`。
2. **选择第 `i` 个元素**：若前 `i - 1` 个元素中存在一个子集的和为 `w - S[i - 1]`（前提是 `w >= S[i - 1]`），那么加上第 `i` 个元素后，子集和就为 `w`，即 `dp[i - 1][w - S[i - 1]]` 为 `True` 时，`dp[i][w]` 也为 `True`。

状态转移方程为：
\[
dp[i][w] = dp[i - 1][w] \lor (dp[i - 1][w - S[i - 1]] \land (w \geq S[i - 1]))
\]

### 时间复杂度分析
需要填充一个大小为 `(n + 1) * (W + 1)` 的二维数组 `dp`，其中 `n` 是集合 `S` 的元素个数，`W` 是目标和。填充每个元素的时间复杂度为 $O(1)$，因此总的时间复杂度为 $O(nW)$。

````python
# add your idea here
def subset_sum(S, W):
    n = len(S)
    # 创建一个二维数组 dp，dp[i][w] 表示前 i 个元素是否存在子集和为 w
    dp = [[False for _ in range(W + 1)] for _ in range(n + 1)]
    # 初始化：当目标和为 0 时，空子集的和为 0，所以 dp[i][0] 都为 True
    for i in range(n + 1):
        dp[i][0] = True
    # 填充 dp 数组
    for i in range(1, n + 1):
        for w in range(1, W + 1):
            # 不选择第 i 个元素
            dp[i][w] = dp[i - 1][w]
            # 如果当前元素小于等于目标和，考虑选择第 i 个元素
            if w >= S[i - 1]:
                dp[i][w] = dp[i][w] or dp[i - 1][w - S[i - 1]]
    # 返回结果
    return dp[n][W]

# 测试示例
S = [1, 4, 7, 3, 5]
W = 11
print(subset_sum(S, W))
    
# your algorithm time complexity is: O(nW)
````

> 给定一个n个物品的集合。物体的重量为$w_1, w_2，…、w_n$，物品的价值分别是$v_1、v_2、…v_n$。给你**两个**重量为 $c$ 的背包。如果你带了一个东西，它可以放在一个背包里，也可以放在另一个背包里，但不能同时放在两个背包里。所有权重和价值都是正整数。

1. 设计一个时间复杂度为 $O(nc^2)$ 的动态规划算法，确定可以放入两个背包的物体的最大价值。分析算法设计思路，并基于python编程实现
2. \* 修改1中的算法，输出每个背包的内容（物品对应下标）。

e.g.: 

输入 V=[1,3,2,5,8,7], W=[1,3,2,5,8,7], c=7, 输出：最大价值=14，背包装的物品为：[6] [4，3] （同一个背包中物品装入顺序对结果无影响）  


idea：
### 算法设计思路

#### 原问题：计算最大价值
可以通过动态规划解决。定义一个三维数组 `dp[i][j][k]` 来表示前 `i` 个物品，第一个背包装载重量为 `j`，第二个背包装载重量为 `k` 时能获得的最大价值。其中 `i` 范围是 `0` 到 `n`（`n` 为物品数量），`j` 和 `k` 范围是 `0` 到 `c`。

对于每个物品 `i`，有以下几种情况：
1. **不放入任何背包**：此时 `dp[i][j][k] = dp[i - 1][j][k]`。
2. **放入第一个背包**：若 `j >= w[i - 1]`，则 `dp[i][j][k] = max(dp[i][j][k], dp[i - 1][j - w[i - 1]][k] + v[i - 1])`。
3. **放入第二个背包**：若 `k >= w[i - 1]`，则 `dp[i][j][k] = max(dp[i][j][k], dp[i - 1][j][k - w[i - 1]] + v[i - 1])`。

最终最大价值为 `dp[n][j][k]` 中的最大值，其中 `0 <= j <= c` 且 `0 <= k <= c`。

#### 修改问题：输出每个背包的内容
在计算最大价值的基础上，我们可以通过回溯的方法来确定每个背包中放入了哪些物品。具体来说，我们可以记录每个状态的选择（不放入、放入第一个背包、放入第二个背包），然后从最终状态开始回溯，找出每个背包中的物品。

### 复杂度分析
由于需要填充一个大小为 `(n + 1) * (c + 1) * (c + 1)` 的三维数组，每次填充操作的时间复杂度为 $O(1)$，所以总的时间复杂度为 $O(nc^2)$。

````python
# add your idea here
def two_knapsacks(V, W, c):
    n = len(V)
    # 初始化三维数组 dp
    dp = [[[0 for _ in range(c + 1)] for _ in range(c + 1)] for _ in range(n + 1)]
    # 记录选择的数组
    choices = [[[0 for _ in range(c + 1)] for _ in range(c + 1)] for _ in range(n + 1)]

    for i in range(1, n + 1):
        for j in range(c + 1):
            for k in range(c + 1):
                # 不放入任何背包
                dp[i][j][k] = dp[i - 1][j][k]
                # 放入第一个背包
                if j >= W[i - 1] and dp[i - 1][j - W[i - 1]][k] + V[i - 1] > dp[i][j][k]:
                    dp[i][j][k] = dp[i - 1][j - W[i - 1]][k] + V[i - 1]
                    choices[i][j][k] = 1
                # 放入第二个背包
                if k >= W[i - 1] and dp[i - 1][j][k - W[i - 1]] + V[i - 1] > dp[i][j][k]:
                    dp[i][j][k] = dp[i - 1][j][k - W[i - 1]] + V[i - 1]
                    choices[i][j][k] = 2

    # 找到最大价值
    max_value = 0
    max_j, max_k = 0, 0
    for j in range(c + 1):
        for k in range(c + 1):
            if dp[n][j][k] > max_value:
                max_value = dp[n][j][k]
                max_j, max_k = j, k

    # 回溯找出每个背包的物品
    bag1 = []
    bag2 = []
    i, j, k = n, max_j, max_k
    while i > 0:
        if choices[i][j][k] == 1:
            bag1.append(i - 1)
            j -= W[i - 1]
        elif choices[i][j][k] == 2:
            bag2.append(i - 1)
            k -= W[i - 1]
        i -= 1

    bag1.sort()
    bag2.sort()
    bag1 = [x + 1 for x in bag1]
    bag2 = [x + 1 for x in bag2]

    print(f"最大价值={max_value}，背包装的物品为：{bag1} {bag2}")

# 测试示例
V = [1, 3, 2, 5, 8, 7]
W = [1, 3, 2, 5, 8, 7]
c = 7
two_knapsacks(V, W, c)
    
# your algorithm time complexity is: O(nc^2)
````

## 问题 10 

> 给定两个字符串 $x[1..n]$ 和 $y[1..m]$，我们想通过以下操作将 $x$ 变换为 $y$ :
>
> **插入**：在 $x$ 中插入一个字符(在任何位置)；**删除**：从 $x$ 中删除一个字符(在任何位置)； **替换**：用另一个字符替换 $x$ 中的一个字符。
>
> 例如: $x = abcd$, $y = bcfe$，
>
> - 将 $x$ 转换为 $y$ 的一种可能方法是：1. 删除 $x$ 开头的 $a$, $x$变成 $bcd$； 2. 将 $x$ 中的字符 $d$ 替换为字符 $f$。$x$ 变成 $bcf$； 3. 在 $x$ 的末尾插入字符 $e$。$x$ 变成 $bcfe$。
>
> - 另一种可能的方法：1. 删除 $x$ 开头的 $a$,  $x$ 变成 $bcd$； 2. 在 $x$ 中字符 $d$ 之前插入字符 $f$。$x$ 变成 $bcfd$。3. 将 $x$ 中的字符 $d$ 替换为字符 $e$。$x$ 变成 $bcfe$。

设计一个时间复杂度为 $O(mn)$ 的算法，返回将 $x$ 转换为 $y$ 所需的最少操作次数。分析算法设计思路，并基于python编程实现。


idea：
### 算法设计思路
本题可使用动态规划来解决两个字符串的编辑距离问题。动态规划的核心思想是将大问题分解为小问题，并保存小问题的解，避免重复计算。

设 `dp[i][j]` 表示将字符串 `x` 的前 `i` 个字符转换为字符串 `y` 的前 `j` 个字符所需的最少操作次数。

- **初始化**：
    - 当 `j = 0` 时，意味着要将字符串 `x` 的前 `i` 个字符转换为空字符串，需要进行 `i` 次删除操作，所以 `dp[i][0] = i`。
    - 当 `i = 0` 时，意味着要将空字符串转换为字符串 `y` 的前 `j` 个字符，需要进行 `j` 次插入操作，所以 `dp[0][j] = j`。

- **状态转移方程**：
    - 若 `x[i - 1] == y[j - 1]`，则不需要进行额外操作，`dp[i][j] = dp[i - 1][j - 1]`。
    - 若 `x[i - 1] != y[j - 1]`，则有三种操作可以选择：
        - **插入**：在 `x` 的前 `i` 个字符后面插入一个字符使其与 `y` 的第 `j` 个字符匹配，此时 `dp[i][j] = dp[i][j - 1] + 1`。
        - **删除**：删除 `x` 的第 `i` 个字符，然后将 `x` 的前 `i - 1` 个字符转换为 `y` 的前 `j` 个字符，此时 `dp[i][j] = dp[i - 1][j] + 1`。
        - **替换**：将 `x` 的第 `i` 个字符替换为 `y` 的第 `j` 个字符，然后将 `x` 的前 `i - 1` 个字符转换为 `y` 的前 `j - 1` 个字符，此时 `dp[i][j] = dp[i - 1][j - 1] + 1`。

    综合这三种操作，取最小值作为 `dp[i][j]` 的值，即 `dp[i][j] = min(dp[i][j - 1] + 1, dp[i - 1][j] + 1, dp[i - 1][j - 1] + 1)`。

- **最终结果**：`dp[n][m]` 即为将字符串 `x` 转换为字符串 `y` 所需的最少操作次数。

### 时间复杂度分析
需要填充一个大小为 `(n + 1) * (m + 1)` 的二维数组 `dp`，其中 `n` 是字符串 `x` 的长度，`m` 是字符串 `y` 的长度。填充每个元素的时间复杂度为 $O(1)$，因此总的时间复杂度为 $O(mn)$。

````python
# add your idea here
def min_edit_distance(x, y):
    n = len(x)
    m = len(y)
    # 创建一个二维数组 dp，dp[i][j] 表示将 x 的前 i 个字符转换为 y 的前 j 个字符所需的最少操作次数
    dp = [[0 for _ in range(m + 1)] for _ in range(n + 1)]

    # 初始化
    for i in range(n + 1):
        dp[i][0] = i
    for j in range(m + 1):
        dp[0][j] = j

    # 填充 dp 数组
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if x[i - 1] == y[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]
            else:
                dp[i][j] = min(dp[i][j - 1] + 1, dp[i - 1][j] + 1, dp[i - 1][j - 1] + 1)

    # 返回结果
    return dp[n][m]

# 测试示例
x = "abcd"
y = "bcfe"
print(f"将字符串 {x} 转换为字符串 {y} 所需的最少操作次数为: {min_edit_distance(x, y)}")
    
# your algorithm time complexity is: O(mn)
````
